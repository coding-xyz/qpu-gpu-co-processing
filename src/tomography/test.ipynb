```python
import math
import itertools
from dataclasses import dataclass
from typing import Dict, Tuple, List

import torch

from tomography import *

device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)

# 1-qubit |+> state: rho = |+><+|
n = 1
ket_plus = (1 / math.sqrt(2)) * torch.tensor([1, 1], device=device, dtype=torch.complex64)
rho_true = ket_plus[:, None] @ ket_plus[None, :].conj()

counts = simulate_counts_from_rho(rho_true, n=n, shots_per_setting=4000, seed=42)

res = mle_tomography(counts, n=n, iters=1500, lr=0.05, device=device, verbose_every=300)
rho_hat = res.rho

print("\nTrue rho:\n", rho_true.detach().cpu().numpy())
print("\nEstimated rho:\n", rho_hat.detach().cpu().numpy())
```
